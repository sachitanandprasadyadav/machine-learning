Is the average total_price of all pizza sales different from $15.

```python
sample_mean=15
pop=df['total_price']
alpha=0.05
```

```python
#(actualmean - sample_mean) / (sd) / root(total_number_of_data)
#import numpy as np

# Calculate the population (actual) mean of total_price
#actual_mean = df['total_price'].mean()

# Calculate the standard deviation of total_price
#sd = df['total_price'].std()

# Calculate the difference between population mean and sample mean
#first_val = actual_mean - sample_mean

# Calculate the standard error: sd / √n
#bot_val = sd / np.sqrt(48619)

# Calculate the t-statistic
#t_statistic = first_val / bot_val

# Display the t-statistic
#t_statistic


```

```python
##stats.ttest_Isamp(population,sample_mean)
t_statistic,p_value=stats.ttest_1samp(pop, popmean=sample_mean)
```

```python
# Print the null hypothesis (H0)
ho='H0: Average total price of all pizza sales is equal to $15'
print(f'{ho}')

# Print the alternative hypothesis (Ha)
ha='Ha: Average total price of all pizza sales is not equal to $15'
print(f'{ha}')

# Print the t-statistic rounded to 2 decimal places
print(f'Test statistic: {t_statistic:,.2f}')

# Print the p-value rounded to 2 decimal places
print(f'P-value: {p_value:,.2f}')

# Print the actual average price of all pizza sales
print(f"Actual average price: Rs. {df['total_price'].mean():,.2f}")

# Print the sample average price
print(f"Sample average price: Rs. {sample_mean:,.2f}")
```

```python
if p_value < alpha:
    print(
        f"Decision: Reject the null hypothesis.\n"
        f"Conclusion: Accept the alternative hypothesis.\n{ha}"
    )
else:
    print(
        f"Decision: Fail to reject the null hypothesis.\n"
        f"Conclusion: Retain the null hypothesis.\n{ho}"
    )
```

Two-sample T-test

the average unit price of classic pizza and chicken pizza are equal?

```python
# Set the significance level (alpha)
alpha = 0.05

# Extract the unit_price values for Classic pizzas
classic_price = df[df['pizza_category'] == 'Classic']['unit_price']

# Extract the unit_price values for Chicken pizzas
chicken_price = df[df['pizza_category'] == 'Chicken']['unit_price']
```

```python
# Syntax:
# stats.ttest_ind(group1, group2, equal_var=False)

t_statistic,p_value=stats.ttest_ind(classic_price, chicken_price, equal_var=False)
```

```python
# Print the null hypothesis (H0)
ho='H0: Average unit price of classic pizza and chicken pizza are equal'
print(f'{ho}')

# Print the alternative hypothesis (Ha)
ha='Ha: Average unit price of classic pizza and chicken pizza are not equal'
print(f'{ha}')

# Print the t-statistic rounded to 2 decimal places
print(f'Test statistic: {t_statistic:,.2f}')

# Print the p-value rounded to 2 decimal places
print(f'P-value: {p_value:,.2f}')

# Print the actual average price of all pizza sales
print(f'Actual classic pizza  price: Rs. {classic_price.mean():,.2f}')
print(f'Actual chicken pizza  price: Rs. {chicken_price.mean():,.2f}')
# Print the sample average price
print(f'Sample average price: Rs. {sample_mean:,.2f}')
```

```python
if p_value < alpha:
    print(
        f"Decision: Reject the null hypothesis.\n"
        f"Conclusion: Accept the alternative hypothesis.\n{ha}"
    )
else:
    print(
        f"Decision: Fail to reject the null hypothesis.\n"
        f"Conclusion: Retain the null hypothesis.\n{ho}"
    )
```

ANOVA Test(Analysis of variance)

Do the average total price generated by all four pizza categories are equal?

```python
categories = df['pizza_category'].unique()
categories
```

```python
grouped_price = []

for category in categories:
    price = df[df['pizza_category'] == category]['total_price']
    grouped_price.append(price)
```

```python
grouped_price
```

```python
# Perform a one-way ANOVA test on the total_price of all pizza categories
t_statistic,p_value=stats.f_oneway(*grouped_price)
```

```python
# Calculate the total price for each pizza category
category_price = df.groupby('pizza_category')['total_price'].sum()

# Format the values with commas and 2 decimal places
format_table = category_price.map("RS.{:,.2f}".format).to_markdown(
    numalign='center',
    stralign='center'
)

# Display the table
print(format_table)
```

```python
ho='ho:Average total price generated by all four pizza are equal.'
ha='ha:Average total price generated by all four pizza are not equal.'
print(ho)
print(ha)
print('-'*36)
print(format_table)
print('-'*36)
print(f't_statistic:{t_statistic}')
print(f'probability value:{p_value}')
```

```python
if p_value < alpha:
    print(
        f"Decision: Reject the null hypothesis.\n"
        f"Conclusion: Accept the alternative hypothesis.\n{ha}"
    )
else:
    print(
        f"Decision: Fail to reject the null hypothesis.\n"
        f"Conclusion: Retain the null hypothesis.\n{ho}"
    )
```
